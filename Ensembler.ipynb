```python
def create_df(row):
    
    if(len(row['passages'])!=10):
        return
    zero_passage_text =''
    
    for dicto2 in row['passages']:
        if(dicto2['is_selected']==0):
            zero_passage_text = dicto2['passage_text']
            break;
    i = 0 
    is_1_found = False
    for dicto1 in row['passages']:
        dicto = {}
        dicto['query_id'] = row['query_id'] 
        dicto['query'] = row['query']
        is_selected = dicto1['is_selected']
        if(is_selected ==1):
            if(is_1_found):
                dicto['label'] = 0
                dicto['passage_text'] = zero_passage_text
            else:
                dicto['label'] = dicto1['is_selected']
                dicto['passage_text'] = dicto1['passage_text']
            is_1_found =True
        else:
            dicto['label'] = dicto1['is_selected']
            dicto['passage_text'] = dicto1['passage_text']
        dicto['passage_id'] = i
        i+=1
        rows.append(dicto)
      
```

```python
import numpy as np
```

```python
def softmax(row):
    row = row.tolist()
#     print (row)
    logits = row[1:]
    exps = [np.exp(i) for i in logits]
    sum_of_exps = sum(exps)
    softmax_ = [j/sum_of_exps for j in exps]
#     print (softmax_)
    return pd.Series(softmax_,index=ans_headers)
```
